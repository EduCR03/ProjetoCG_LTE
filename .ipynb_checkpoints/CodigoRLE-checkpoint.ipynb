```python
import matplotlib.pyplot as plt
import moviepy.editor as mpe
import numpy as np

video = mpe.VideoFileClip("carphone.mp4")
```

```python
# Extração dos frames 72 e 73:
frame_number = 72
frame0 = video.get_frame(frame_number / video.fps)
frame1 = video.get_frame((frame_number + 1) / video.fps)
```

```python
# Diferença entre imagens:
frame_diff = frame1.astype(np.int16) - frame0.astype(np.int16)

# Quantização:
delta = 10
frame_quantized = (delta * np.floor(frame_diff / delta + 0.5)).astype(np.int16)
```

```python
plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.title("Frame0")
plt.imshow(frame0)

plt.subplot(2, 2, 2)
plt.title("Frame1")
plt.imshow(frame1)

plt.subplot(2, 2, 3)
plt.title("Diferença entre Frames")
plt.imshow(frame_diff)

plt.subplot(2, 2, 4)
plt.title("Quantizado")
plt.imshow(frame_quantized)
```

```python
def rle_encode(data):
    
    # lista vazia
    encoding = []
    
    # se 'data' não tiver nada retorna lista vazia
    if data.size == 0:
        return encoding
    
    # pega o elemento 0
    prev = data[0]
    count = 0
    
    # para cada valor em 'data'
    for value in data:
        # se for diferente de 'data[0]'
        if value != prev:
            # adiciona o contador
            encoding.append(count)
            # adiciona data[0]
            encoding.append(prev)
    
            
            count = 1
            # prev recebe 'value'
            prev = value
        else:
            # caso não, contador aumenta em 1
            count += 1

    # adiciona o 'count' e o 'prev' na lista
    encoding.append(count)
    encoding.append(prev)
    
    # retorna o array
    return np.array(encoding,dtype=np.int16)
```

```python
frame_rle = rle_encode(frame_quantized.flatten())

print("Tamanho Original:", frame0.nbytes)
print("Tamanho Comprimido:", frame_rle.nbytes)
print("Taxa de Compressão:", frame0.nbytes / frame_rle.nbytes)
```

```python
def rle_decode(data):
    # lista vazia
    output = []
    is_repeat = True
    
    # para cada valor em 'data'
    for value in data:
        # se for 'true'
        if is_repeat:
            # repeat recebe 'value'
            repeat = value
        else:
            # caso não, adiciona 'value' na lista
            for _ in range(repeat):
                output.append(value)
        
        # vira falso
        is_repeat = not is_repeat

    # retorna o array
    return np.array(output,dtype=np.int16)
```

```python
# realoca o tamanho para o mesmo do frame0
frame_decoded = np.reshape(rle_decode(frame_rle), frame0.shape)
# faz a soma
frame_sum = (frame_decoded + frame0).astype(np.uint8)

plt.title("Frame0 Descomprimido")
plt.imshow(frame_sum)
```

```python
print("Tamanho Descomprimido:", frame_sum.nbytes)
```
